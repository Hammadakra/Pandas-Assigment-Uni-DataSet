### 11. How many students got A, B, C, F?

```python
countA=0
countB=0
countC=0
countD=0

for i in range(len(df)):
    if df.Grade[i] == 'A':
        countA+=1
    elif df.Grade[i] == 'B':
        countB+=1
    elif df.Grade[i] == 'C':
        countC+=1
    elif df.Grade[i] == 'F':
        countD+=1
          
print("Number of Student Got A  = ",countA)
print("Number of Student Got B  = ",countB)
print("Number of Student Got C  = ",countC)
print("Number of Student Got D  = ",countD)
```
